```python
from alpha_vantage.timeseries import TimeSeries
import pandas as pd
import time

# Define your API key
api_key = "DZN0SFNNTOZKAY20"

# Initialize the Alpha Vantage TimeSeries object
ts = TimeSeries(key=api_key, output_format='pandas')

# Function to fetch data in chunks
def fetch_monthly_minute_data(symbol, start_date, end_date):
    """
    Fetch 1-minute data for a stock over a month using Alpha Vantage API.

    Args:
        symbol (str): Stock ticker (e.g., "AAPL").
        start_date (str): Start date in YYYY-MM-DD format.
        end_date (str): End date in YYYY-MM-DD format.

    Returns:
        DataFrame: Combined 1-minute data over the specified range.
    """
    all_data = []
    current_date = pd.to_datetime(start_date)

    while current_date <= pd.to_datetime(end_date):
        # Fetch 1-minute intraday data (compact mode)
        try:
            print(f"Fetching data for {current_date.strftime('%Y-%m-%d')}...")
            data, metadata = ts.get_intraday(symbol=symbol, interval="1min", outputsize="full")
            all_data.append(data)

            # Wait to respect API limits (5 requests per minute for free tier)
            time.sleep(12)  # Add delay to avoid exceeding request limits
        except Exception as e:
            print(f"Error fetching data for {current_date.strftime('%Y-%m-%d')}: {e}")

        # Move to the next day
        current_date += pd.Timedelta(days=1)

    # Combine all dataframes into one
    if all_data:
        full_data = pd.concat(all_data)
        full_data.sort_index(inplace=True)
        return full_data
    else:
        return None

# Example usage: Fetch data for AAPL from 2025-01-01 to 2025-01-31
start_date = "2025-01-01"
end_date = "2025-01-31"
symbol = "INFY"

data = fetch_monthly_minute_data(symbol, start_date, end_date)

# Save to a CSV file for later use
if data is not None:
    data.to_csv(f"{symbol}_1min_{start_date}_to_{end_date}.csv")
    print(f"Data saved to {symbol}_1min_{start_date}_to_{end_date}.csv")
else:
    print("No data fetched.")
```

```python
ema=df['4. close'].head(3).mean()
median_sma = ema
df['EMA'] = None  # Initialize SMA column
df.loc[:2, 'SMA'] = median_sma  # Fill first 3 rows with the median value
n = 5  # EMA period
df['EMA'] = df['4. close'].ewm(span=n, adjust=False).mean()
df = df.drop('SMA', axis=1)
```

```python
df
```

```python
short_time_period=1
long_time_period=60
df['Short_Volume_MA'] = df['5. volume'].rolling(window=short_time_period, min_periods=1).mean()
df['Long_Volume_MA'] = df['5. volume'].rolling(window=long_time_period, min_periods=1).mean()
df['Volume_Oscillator'] = ((df['Short_Volume_MA'] - df['Long_Volume_MA']) / df['Long_Volume_MA']) * 100

# Drop the intermediate moving average columns if you don't need them
df.drop(['Short_Volume_MA', 'Long_Volume_MA'], axis=1, inplace=True)
```

```python
n=15 #minutes
df['price change']=df['4. close'].diff()
df['gain']=df['price change'].apply(lambda x: x if x>0 else 0)
df['Loss'] = df['price change'].apply(lambda x: -x if x < 0 else 0)
df['Avg Gain'] = df['gain'].rolling(window=n, min_periods=1).mean()
df['Avg Loss'] = df['Loss'].rolling(window=n, min_periods=1).mean()
df['RS'] = df['Avg Gain'] / df['Avg Loss']
df['RSI'] = 100 - (100 / (1 + df['RS']))
df.drop(['price change', 'gain', 'Loss', 'Avg Gain', 'Avg Loss', 'RS'], axis=1, inplace=True)
```

```python
df
```

```python
n=15
m=3
df['Lowest Low'] = df['3. low'].rolling(window=n, min_periods=1).min()
df['Highest High'] = df['2. high'].rolling(window=n, min_periods=1).max()
df['%K'] = ((df['4. close'] - df['Lowest Low']) / (df['Highest High'] - df['Lowest Low'])) * 100
df['%D'] = df['%K'].rolling(window=m, min_periods=1).mean()
df.drop(['Lowest Low', 'Highest High'], axis=1, inplace=True)
```

```python
df
```

```python
import numpy as np
period = 15
df['Previous Close'] = df['4. close'].shift(1)
df['TR'] = np.maximum(
    df['2. high'] - df['3. low'],
    np.maximum(
        abs(df['2. high'] - df['Previous Close']),
        abs(df['3. low'] - df['Previous Close'])
    )
)
df['+DM'] = np.where(
    (df['2. high'] - df['2. high'].shift(1)) > (df['3. low'].shift(1) - df['3. low']),
    np.maximum(df['2. high'] - df['2. high'].shift(1), 0),
    0
)
df['-DM'] = np.where(
    (df['3. low'].shift(1) - df['3. low']) > (df['2. high'] - df['2. high'].shift(1)),
    np.maximum(df['3. low'].shift(1) - df['3. low'], 0),
    0
)
df['Smoothed_TR'] = df['TR'].rolling(window=period, min_periods=1).sum()
df['Smoothed_+DM'] = df['+DM'].rolling(window=period, min_periods=1).sum()
df['Smoothed_-DM'] = df['-DM'].rolling(window=period, min_periods=1).sum()
df['+DI'] = (df['Smoothed_+DM'] / df['Smoothed_TR']) * 100
df['-DI'] = (df['Smoothed_-DM'] / df['Smoothed_TR']) * 100
df['DX'] = (abs(df['+DI'] - df['-DI']) / (df['+DI'] + df['-DI'])) * 100

# Step 6: Calculate the Average Directional Index (ADX)
df['ADX'] = df['DX'].rolling(window=period, min_periods=1).mean()

# Drop unnecessary columns
df.drop(['Previous Close', 'TR', '+DM', '-DM', 'Smoothed_TR', 'Smoothed_+DM', 'Smoothed_-DM', 'DX'], axis=1, inplace=True)
```

```python
df
```

```python
df['Percentage Price Change'] = df['4. close'].pct_change()

# Step 2: Calculate the Price Volume Trend (PVT)
df['PVT'] = (df['Percentage Price Change'] * df['5. volume']).cumsum()

# Drop unnecessary columns
df.drop(['Percentage Price Change'], axis=1, inplace=True)
```

```python
df['Target'] = df['4. close'].pct_change()
```

```python
df
```

```python
df['previous_close'] = df['4. close'].shift(1)
df['TR'] = df[['2. high', '3. low', 'previous_close']].apply(
    lambda row: max(row['2. high'] - row['3. low'], abs(row['2. high'] - row['previous_close']), abs(row['3. low'] - row['previous_close'])), axis=1)
df['ATR'] = df['TR'].rolling(window=14).mean()
df['Max_Gain'] = df['2. high'] - df['1. open']
df['Max_Loss'] = df['1. open'] - df['3. low']
df['RRR_ATR'] = df['Max_Gain'] / df['ATR']
df['Returns'] = df['4. close'].pct_change()
risk_free_rate = 0.015
df['Excess_Returns'] = df['Returns'] - risk_free_rate / 252
df['Downside_Returns'] = np.where(df['Excess_Returns'] < 0, df['Excess_Returns'], 0)
downside_deviation = np.sqrt((df['Downside_Returns'] ** 2).mean())
sortino_ratio = df['Excess_Returns'].mean() / downside_deviation

# Combine ATR and Sortino Ratio for comprehensive RRR
df['Comprehensive_RRR'] = df['RRR_ATR'] * sortino_ratio
```

```python
comprehensive_RRR= df['Comprehensive_RRR'].median()

# Fill NaN values in the 'RSI' column with the mean value
df['Comprehensive_RRR'].fillna(comprehensive_RRR, inplace=True)
```

```python
rsi_mean = df['RSI'].mean()

# Fill NaN values in the 'RSI' column with the mean value
df['RSI'].fillna(rsi_mean, inplace=True)
percent_changeink = df['%K'].median()
df['%K'].fillna(percent_changeink, inplace=True)

percent_changeind=df['%D'].median()
df['%D'].fillna(percent_changeind, inplace=True)

plus_d=df['+DI'].median()
df['+DI'].fillna(plus_d, inplace=True)

minus_d=df['-DI'].median()
df['-DI'].fillna(minus_d, inplace=True)

adx=df['ADX'].median()
df['ADX'].fillna(adx, inplace=True)

pvt=df['PVT'].median()
df['PVT'].fillna(pvt, inplace=True)

target=df['Target'].median()
df['Target'].fillna(target, inplace=True)
```
